# Decade Seperation

```python
import pandas as pd
import re, math
from utils.constants import Paths
```

```python
# Load excel that had rows w/o transcrips pulled out
print(Paths.mccray_with_trscp)
df = pd.read_excel(Paths.mccray_with_trscp)
print(list(df))
df

```

```python
# Isolate the data given a specific range of years
start_year = 1941
end_year = 1950

df_1940s = df[df['Year'] != ""].copy() 
df_1940s['Year'] = df_1940s['Year'].apply(lambda x: 0 if math.isnan(x) else int(x))
df_1940s = df_1940s[df_1940s['Year'] >= start_year]
df_1940s = df_1940s[df_1940s['Year'] <= end_year]
df_1940s
```

```python
# Save and print trscps
df_1940s.to_excel(Paths.mccray_1940s_subset)
df_1940s['Transcript']
```

```python
# remove messy_char != 0 and save
df_1940_cleanish = df_1940s[df_1940s['any_messy'] == False]
df_1940_cleanish.to_excel(Paths.mccray_1940s_cleaner)
```

```python
# get messy_char to share
df_1940_unclean = df_1940s[df_1940s['any_messy'] == True]
df_1940_unclean.to_excel(Paths.mccray_folder + Paths.mccray_1940s_unclean)
```
