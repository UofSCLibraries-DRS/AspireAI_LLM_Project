Decade Seperation

```python
import pandas as pd
import re
import ftfy
# https://ftfy.readthedocs.io/en/v6.3.0/explain.html
```

```python
df = pd.read_excel("McCray metadata (with transcripts).xlsx")
df
```

```python
print(df['Date'])
```

```python
# https://www.geeksforgeeks.org/pandas/convert-a-dataframe-column-to-integer-in-pandas/

df['Year'] = df['Date']
df['Year'] = df['Year'].apply(lambda x: str(x))
df['Year'] = df['Year'].apply(lambda x: re.sub(r'[^0-9]+', "", x))
df['Year'] = df['Year'].apply(lambda x: x[:4])
df['Year']
```

```python
df.to_excel('McCray (with transcripts + years).xlsx', index=False)
```

```python
start_year = 1940
end_year = 1950

df_1940s = df[df['Year'] != ""].copy() 
df_1940s['Year'] = df_1940s['Year'].apply(lambda x: int(x))
df_1940s = df_1940s[df_1940s['Year'] >= start_year]
df_1940s = df_1940s[df_1940s['Year'] <= end_year]
df_1940s
```

```python
df_1940s.to_excel('McCray (1940-1950).xlsx', index=False)
df_1940s['Transcript']
```

```python
df_chars = df_1940s.copy()
df_chars = df_chars[['Title', 'CONTENTdm number', 'Year', 'Transcript']]
df_chars['Messy Chars'] = df_chars['Transcript'].apply(lambda x: re.sub(r'[a-zA-Z0-9$.,!?:"()\s]+', "", x))
df_chars['Err Count'] = df_chars['Messy Chars'].str.len()

df_chars['Cleaned'] = df_chars['Transcript'].copy()
df_chars['Cleaned'] = df_chars['Cleaned'].apply(lambda x: re.sub(r'[^a-zA-Z0-9$.,!?:"()\s]+', "", x))
df_chars['Cleaned'] = df_chars['Cleaned'].apply(lambda x: re.sub(r'\s{ 2,}', " ", x))
df_chars['Cleaned'] = df_chars['Cleaned'].apply(lambda x: ftfy.fix_text(x))

df_chars.to_excel('McCray messy chars review (new).xlsx', index=False)
df_chars
```
