```python
df = pd.read_excel(subset_file_path)

print(list(df))

df["Clean Transcript"] = df["Clean Transcript"].str.replace(
    r"^1022 W",   # match only beginning "1022 W"
    "1022½ W",
    regex=True
)

df
```

```python
cols = [
    "CONTENTdm number",
    "Reference URL",
    "Title",
    "Description",
    "Year",
    "Clean Transcript"
]

df = df[cols]

df
```

```python
df_og = pd.read_csv(r'/home/steven-dindl/Downloads/McCray+ (Test2, All Fields).csv')

# if df['CONTENTdm number'] (this is the subset) == df_og['CONTENTdm number'] (inside the larger main data) then add row to df_all
df_all = df_og[df_og['CONTENTdm number'].isin(df['CONTENTdm number'])]

cols = [
    "CONTENTdm number",
    "Reference URL",
    "Title",
    "Description",
    "Year",
    "Original Transcript",
    "Semi-clean Transcript"
]

df_all = df_all[cols]

# all together (adding 'Clean Transcripts' by merging on 'CONTENTdm number')
df_all = df_all.merge(
    df[['CONTENTdm number', 'Clean Transcript']], 
    on='CONTENTdm number', 
    how='left'
)

df_OG_trsp = df_all.drop(["Semi-clean Transcript", "Clean Transcript"], axis=1)
df_OG_trsp = df_OG_trsp.rename(columns={'Original Transcript': 'Transcript'})

df_semi_trsp = df_all.drop(["Original Transcript", "Clean Transcript"], axis=1)
df_semi_trsp = df_semi_trsp.rename(columns={'Semi-clean Transcript': 'Transcript'})

df = df_all.drop(["Original Transcript", "Semi-clean Transcript"], axis=1)
df = df.rename(columns={'Clean Transcript': 'Transcript'})

display(df_all)
display(df)
display(df_OG_trsp)
display(df_semi_trsp)
```

```python
def count_empty_column(df, column_name):
    print(f"{column_name} empty record count: {df[column_name].isnull().sum() + (df[column_name].astype(str).str.strip() == '').sum()}")

count_empty_column(df_all, "Original Transcript")
count_empty_column(df_all, "Semi-clean Transcript")
count_empty_column(df_all, "Clean Transcript")
```
